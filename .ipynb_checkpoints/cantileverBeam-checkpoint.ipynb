| Axis | What it Represents | How It's determined|
|----------|----------|----------|
| local x | The centerline of the beam | Line connection Node I to Node J|
| local y | The vertical axis of the cross-section | Vy = Vxz * Vx|
| local z | The Horizontal axis of the cross section | Vz = Vx * Vy|
| Global X | Horizontal (left to right)||
|Global Y | Vertical (upward)||
| Global Z | Points out of the screen||

|  Type | Descriptioni | When to use | 
|----------|----------|----------|
| Linear | Assumes small displacements. The geometry is considered unchanged during the analysis. | Standard static or linear dynamic analysis where deformation are very small|
| PDelta | Accounts for $P-\Delta$ effects ( second order effects.) It considers the effect of axial loads on the lateral stiffness.| Tall buildings or slender column where axial loads significantly affect bending|
|Corotational | Supports large displacements and large rotations (but small strains). It "rotates" the local coordinate system with the element. | Large-scale buckling, cables, or structures undergoing extreme deformation|

```python
ops.geomTransf('Linear', 1)
```

| Element Category | Element Name | Best Used For | Key Characteristics|
|----------|----------|----------|----------|
|Truss | truss | Braces, bridge stay cable, and simple axial links | Only carries axial force (tension or compresssion) no bending|
|| corotTruss| Large-displacement cables or slender brace pron to buckling | Accounts for large geometric changes (corotational formulation)|
| Beam-Column | elasticBeamColumn | Preliminary Frame analysis, steel members under small loads. |Fast computation, assumes the material never yields or breaks.|
|| forceBeamColumn| Earthquake engineering, collapse simulation, plastic hinges. | Nonlinear. Uses fiber section to track yielding and damage along the length.|
|| dispBeamColumn | Elemnts with significant non-uniform loads or specific curvature.| Displacement-based formulation often requires a finer mesh than force-based.|
| Zero-Length | zeroLength | Modeling pins, bolts, soil springs, or seismic isolators. | Connects two nodes at the same coordinate. Uses UniaxialMaterial to define behavior.|
|Area/Shell | shellMITC4 | Floor slabs, shear walls, and storage tanks. | Handles both in-plane (membrance) and out-of-plane(bending forces).|
||quad | 2D plane stress/strian problems (eg. cross section of a dam)| Used for 2D continuum modeling|
| Solid | stdBrick | Complex 3D connections, heavy mass concrete, or soil blocks.| 8-nodes 3D element for full volumetric stress analysis|

| Test Name | What is Monitors | Why use it? |
|---------|----------|---------|
|NormDispIncr | change in Displacement | Good for simple models; focuses on "settling" into a position|
|NormUnbalance | Remaining Force Error | Most common for structural engineering; ensures F = ma (equilibrium)|
| EnergyIncr | Force * Displacement | Best for complex models with rotations and translations. |
| RelativeNormUnbalance | Force Error relative to load | Good when your loads are very small or very large (scales the tolorance).|

```python
# ops.test(method, tolerance, max_iteration)
ops.test('NormDispIncr', 1.0e-6, 10)
```

```python
ops.analyze(1)
```

By using Formula to get the displacement 

$$\delta = \frac{PL^3}{3EI}$$

```python
#ops.nodeDisp(nodeTag, index) ---> return array of forces in x, y and moment if index not provided
displacement = ops.nodeDisp(2, 2)
calc = (yF * L**3)/(3 * E * Iz)
```

```python
print(f"The computed displacement is {displacement}")
print(f"The mathematically obtained displacement is {calc}")
```
